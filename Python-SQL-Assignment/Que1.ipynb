```python
class DB:
    def __init__(self,host,user,password):  
        try:
            self.db_conn = mysql.connector.connect(host=host,user=user,password=password)        
        except Exception as e:
            print(e)
            sys.exit(0)
            
    def create_db(self):
        try:    
            query_delete_if_db_exists = "DROP DATABASE IF EXISTS ecomm_db_"
            query_create_db = "CREATE DATABASE ecomm_db_"
            query_use_db = "USE ecomm_db_"
            db_cursor = self.db_conn.cursor()        
            db_cursor.execute(query_delete_if_db_exists) 
            db_cursor.execute(query_create_db) 
            db_cursor.execute(query_use_db) 
        except Exception as e:
            print(e)
            sys.exit(0)
            
            
    def create_schema_for_tables(self):
        try:
            query_user_table = "CREATE TABLE users (user_id int NOT NULL AUTO_INCREMENT,created_at datetime NOT NULL DEFAULT CURRENT_TIMESTAMP,email varchar(100) NOT NULL,password varchar(100) NOT NULL,full_name varchar(100) DEFAULT NULL,phone_no decimal(10,0) DEFAULT NULL,reside_city varchar(100) DEFAULT NULL,PRIMARY KEY (user_id))"
            query_retailer_table = "CREATE TABLE retailers (retailer_id int NOT NULL AUTO_INCREMENT,full_name varchar(100) DEFAULT NULL,email varchar(100) NOT NULL,operation_city varchar(100) NOT NULL,PRIMARY KEY (retailer_id))"
            query_products_table = "CREATE TABLE products (product_id int NOT NULL AUTO_INCREMENT,created_at datetime NOT NULL DEFAULT CURRENT_TIMESTAMP,product_name varchar(100) DEFAULT NULL,by_retailer_id int NOT NULL,manufactured_at datetime DEFAULT NULL,description varchar(1000) DEFAULT NULL,price decimal(20,2) DEFAULT NULL,category enum('novel','electronic','cloth') NOT NULL,PRIMARY KEY (product_id),  FOREIGN KEY (by_retailer_id) REFERENCES retailers (retailer_id))"
            query_product_views = "CREATE TABLE product_views (view_id int NOT NULL AUTO_INCREMENT,by_user_id int NOT NULL,for_product_id int NOT NULL,created_at datetime NOT NULL DEFAULT CURRENT_TIMESTAMP,PRIMARY KEY (view_id),FOREIGN KEY (by_user_id) REFERENCES users (user_id) ON DELETE CASCADE,  FOREIGN KEY (for_product_id) REFERENCES products (product_id))"
            query_rating_and_reviews = "CREATE TABLE ratings_and_reviews (rating_id int NOT NULL AUTO_INCREMENT,by_user_id int DEFAULT NULL,display_full_name varchar(100) DEFAULT NULL,rating_score enum('1','2','3','4','5') NOT NULL,review_text varchar(1000) DEFAULT NULL,for_product_id int NOT NULL,created_at datetime NOT NULL DEFAULT CURRENT_TIMESTAMP,PRIMARY KEY (rating_id),  FOREIGN KEY (by_user_id) REFERENCES users (user_id),FOREIGN KEY (for_product_id) REFERENCES products (product_id))"
            query_orders = "CREATE TABLE orders (order_id int NOT NULL AUTO_INCREMENT,by_user_id int NOT NULL,order_status enum('in_cart','placed') NOT NULL,created_at datetime NOT NULL DEFAULT CURRENT_TIMESTAMP,PRIMARY KEY (order_id),  FOREIGN KEY (by_user_id) REFERENCES users (user_id) ON DELETE CASCADE)"
            query_order_product_details = "CREATE TABLE order_product_details (product_id int NOT NULL,quantity int NOT NULL,of_order_id int NOT NULL,PRIMARY KEY (product_id,of_order_id),  FOREIGN KEY (product_id) REFERENCES products (product_id) ON DELETE CASCADE,FOREIGN KEY (of_order_id) REFERENCES orders (order_id) ON DELETE CASCADE)"
            query_delivery_addresses = "CREATE TABLE delivery_addresses (address_id int NOT NULL AUTO_INCREMENT,of_user_id int NOT NULL,building_no varchar(100) DEFAULT NULL,street_name varchar(100) DEFAULT NULL,locality varchar(100) DEFAULT NULL,city varchar(100) NOT NULL,state varchar(100) NOT NULL,country varchar(100) NOT NULL,pincode int NOT NULL,PRIMARY KEY (address_id),FOREIGN KEY (of_user_id) REFERENCES users (user_id) ON DELETE CASCADE)"
            query_novel_details = "CREATE TABLE novel_details (detail_id int NOT NULL AUTO_INCREMENT,author_name varchar(200) NOT NULL,publish_date datetime NOT NULL,for_product_id int NOT NULL,PRIMARY KEY (detail_id),FOREIGN KEY (for_product_id) REFERENCES products (product_id))"
            db_cursor = self.db_conn.cursor()        
            db_cursor.execute(query_user_table)
            db_cursor.execute(query_retailer_table)
            db_cursor.execute(query_products_table)
            db_cursor.execute(query_product_views)
            db_cursor.execute(query_rating_and_reviews)
            db_cursor.execute(query_orders)
            db_cursor.execute(query_order_product_details)
            db_cursor.execute(query_delivery_addresses)
            db_cursor.execute(query_novel_details)
        except Exception as e:
            print(e)
            sys.exit(0)            
            
    def Que1(self, user_data, retailer_data, products_data, ratings_and_reviews_data, product_views_data, orders_data, order_product_details_data, novel_details_data, delivery_addresses_data):
        try:
            db_cursor = self.db_conn.cursor()
                        
            query = "INSERT INTO users VALUES (%s, %s, %s, %s, %s, %s, %s)"
            db_cursor.executemany(query, user_data)
            self.db_conn.commit()
                
            
            query = "INSERT INTO retailers VALUES (%s, %s, %s, %s)"
            db_cursor.executemany(query, retailer_data)
            self.db_conn.commit()
                
            query = "INSERT INTO products VALUES (%s, %s, %s, %s, %s, %s, %s, %s)"
            db_cursor.executemany(query, products_data)
            self.db_conn.commit()
                            
            query = "INSERT INTO ratings_and_reviews VALUES (%s, %s, %s, %s, %s, %s, %s)"
            db_cursor.executemany(query, ratings_and_reviews_data)
            self.db_conn.commit()
                            
            query = "INSERT INTO product_views VALUES (%s, %s, %s, %s)"
            db_cursor.executemany(query, product_views_data)
            self.db_conn.commit()
                            
            query = "INSERT INTO orders VALUES (%s, %s, %s, %s)"
            db_cursor.executemany(query, orders_data)
            self.db_conn.commit()
                            
            query = "INSERT INTO order_product_details VALUES (%s, %s, %s)"
            db_cursor.executemany(query, order_product_details_data)
            self.db_conn.commit()
                            
            query = "INSERT INTO novel_details VALUES (%s, %s, %s, %s)"
            db_cursor.executemany(query, novel_details_data)
            self.db_conn.commit()
                            
            query = "INSERT INTO delivery_addresses VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)"
            db_cursor.executemany(query, delivery_addresses_data)
            self.db_conn.commit()                                            
            
        except Exception as e:
            print(e)
            sys.exit(0)
            
    def Que2(self, user_id):
        try:
            query_1 = "UPDATE ratings_and_reviews SET by_user_id = NULL, display_full_name = 'Anonymous' WHERE by_user_id = %s"
            query_2 = "DELETE FROM users WHERE user_id = %s"
            db_cursor = self.db_conn.cursor()        
            db_cursor.execute(query_1,(user_id,))  
            self.db_conn.commit()
            print(db_cursor.rowcount, "record(s) affected")    
            db_cursor.execute(query_2,(user_id,)) 
        except Exception as e:
            print(e)
            sys.exit(0)
            
    def Que3(self,amount,percent_increase,views,months):
        try:
            query = "UPDATE products SET price = %s*price WHERE (product_id IN (SELECT for_product_id FROM (SELECT for_product_id, COUNT(*) as views_for_product_id FROM product_views WHERE created_at > DATE_SUB(CURRENT_TIMESTAMP() , INTERVAL %s MONTH) GROUP BY for_product_id) as product_ids_to_edit WHERE views_for_product_id>=%s)) AND price < %s"
            db_cursor = self.db_conn.cursor()        
            db_cursor.execute(query,(1+(percent_increase/100),months,views,amount,))  
            self.db_conn.commit()
            print(db_cursor.rowcount, "record(s) affected")      
        except Exception as e:
            print(e)
            sys.exit(0)
        
    def Que5(self, city_name, purchase_amount):
        try:
            query = "SELECT users.phone_no, users.email FROM orders INNER JOIN order_product_details ON orders.order_id = order_product_details.of_order_id INNER JOIN products ON products.product_id = order_product_details.product_id INNER JOIN users ON users.user_id = orders.by_user_id WHERE orders.order_status = 'placed' AND users.reside_city = %s GROUP BY orders.by_user_id HAVING SUM(order_product_details.quantity*products.price) >= %s;"        
            db_cursor = self.db_conn.cursor()        
            db_cursor.execute(query, (city_name, purchase_amount, ))        
            column_names = [column[0] for column in db_cursor.description]        
            data = db_cursor.fetchall()        
            self.print_for_select(column_names,data)            
        except Exception as e:
            print(e)
            sys.exit(0)
            
    def Que6(self,containing_string):
        try:
            query = "SELECT desired_products.product_id as product_id,desired_products.created_at as product_created_at,desired_products.product_name as product_name,desired_products.by_retailer_id as by_retailer_id,desired_products.manufactured_at as manufactured_at,desired_products.description as description,desired_products.price as price, desired_products.category as category,users_who_bought_atleast_once.user_id as user_id,users_who_bought_atleast_once.created_at as user_created_at,users_who_bought_atleast_once.email as email,users_who_bought_atleast_once.password as password,users_who_bought_atleast_once.full_name as full_name,users_who_bought_atleast_once.phone_no as phone_no,users_who_bought_atleast_once.reside_city as reside_city FROM (SELECT * FROM products WHERE products.product_name LIKE %s) as desired_products LEFT OUTER JOIN (SELECT DISTINCT users.*, order_product_details.product_id FROM orders INNER JOIN order_product_details ON orders.order_id = order_product_details.of_order_id INNER JOIN users ON users.user_id = orders.by_user_id WHERE orders.order_status = 'placed' AND order_product_details.product_id IN (SELECT product_id FROM products WHERE products.product_name LIKE %s)) as users_who_bought_atleast_once ON desired_products.product_id = users_who_bought_atleast_once.product_id;"        
            db_cursor = self.db_conn.cursor()        
            db_cursor.execute(query, ("%"+containing_string+"%", "%"+containing_string+"%"))
            column_names = [column[0] for column in db_cursor.description]        
            data = db_cursor.fetchall()        
            self.print_for_select(column_names,data)            
        except Exception as e:
            print(e)
            sys.exit(0)
    
    def Que13(self):
        try:
            query = "SELECT *  FROM products WHERE (category = 'electronic') AND ((product_name LIKE '%laptop%') OR (description LIKE '%laptop%')) ORDER BY price ASC;"
            db_cursor = self.db_conn.cursor()        
            db_cursor.execute(query)
            column_names = [column[0] for column in db_cursor.description]        
            data = db_cursor.fetchall()        
            self.print_for_select(column_names,data)            
        except Exception as e:
            print(e)
            sys.exit(0)
        
    def Que16(self, quantity):
        try:
            query = "SELECT DISTINCT users.user_id, users.phone_no, users.email FROM orders INNER JOIN order_product_details ON orders.order_id = order_product_details.of_order_id INNER JOIN users ON users.user_id = orders.by_user_id WHERE orders.order_status = 'in_cart'  GROUP BY orders.order_id HAVING MIN(order_product_details.quantity) < %s;"
            db_cursor = self.db_conn.cursor()        
            db_cursor.execute(query, (quantity, ))
            column_names = [column[0] for column in db_cursor.description]        
            data = db_cursor.fetchall()        
            self.print_for_select(column_names,data)            
        except Exception as e:
            print(e)
            sys.exit(0)
        
    def Que19(self, user_id):
        try:
            query = "SELECT DISTINCT products.by_retailer_id as retailer_id FROM orders INNER JOIN order_product_details ON orders.order_id = order_product_details.of_order_id INNER JOIN products ON products.product_id = order_product_details.product_id WHERE (orders.by_user_id = %s) AND (orders.order_status = 'placed');"
            db_cursor = self.db_conn.cursor()        
            db_cursor.execute(query, (user_id, ))
            column_names = [column[0] for column in db_cursor.description]        
            data = db_cursor.fetchall()        
            self.print_for_select(column_names,data)            
        except Exception as e:
            print(e)
            sys.exit(0)
        
    def Que20(self,discount_percent):
        try:
            query_1 = "CREATE TABLE IF NOT EXISTS holi_Deals SELECT product_id, created_at, product_name, by_retailer_id, manufactured_at, description, %s*price as price, category FROM products WHERE created_at >= DATE_SUB(CURRENT_TIMESTAMP(), INTERVAL 100 DAY)"
            query_2 = "UPDATE products SET price = %s*price WHERE created_at >= DATE_SUB(CURRENT_TIMESTAMP(), INTERVAL 100 DAY)"
            db_cursor = self.db_conn.cursor() 
            db_cursor.execute(query_1, ((100-discount_percent)/100, ))               
            db_cursor.execute(query_2, ((100-discount_percent)/100, ))    
            self.db_conn.commit()  
            print(db_cursor.rowcount, "record(s) affected")
            
        except Exception as e:            
            print(e)
            sys.exit(0)
        
    def Que21(self, user_id):
        try:
            query_1 = "DROP TABLE IF EXISTS product_scores_by_users"
            query_2 = "CREATE TABLE IF NOT EXISTS product_scores_by_users (by_user_id INT NOT NULL,for_product_id INT NOT NULL,total_quantity_purchased INT,  total_views INT,  product_score DECIMAL(20,15) NOT NULL,    avg_product_score DECIMAL(20,15) NOT NULL,final_product_score DECIMAL(20,15) NOT NULL,PRIMARY KEY (by_user_id, for_product_id))"
            query_3 = "INSERT INTO product_scores_by_users SELECT users.user_id as by_user_id, products.product_id as for_product_id, 0, 0, 0.0, 0.0, 0.0 FROM users, products"
            query_4 = "UPDATE product_scores_by_users SET total_quantity_purchased = (SELECT SUM(order_product_details.quantity)FROM orders INNER JOIN order_product_details ON orders.order_id = order_product_details.of_order_id WHERE orders.by_user_id = product_scores_by_users.by_user_id AND order_product_details.product_id = product_scores_by_users.for_product_id AND orders.order_status = 'placed')";
            query_5 = "UPDATE product_scores_by_users SET total_quantity_purchased = 0 WHERE total_quantity_purchased IS NULL"
            query_6 = "UPDATE product_scores_by_users SET total_views = (SELECT COUNT(*) FROM product_views WHERE product_views.by_user_id = product_scores_by_users.by_user_id AND product_views.for_product_id = product_scores_by_users.for_product_id)"
            query_7 = "UPDATE product_scores_by_users SET product_score = total_quantity_purchased*0.8 + total_views*0.2"
            query_8 = "UPDATE product_scores_by_users SET avg_product_score = (SELECT AVG(B.product_score) FROM (SELECT * FROM product_scores_by_users) as B WHERE B.by_user_id = product_scores_by_users.by_user_id GROUP BY B.by_user_id)"
            query_9 = "UPDATE product_scores_by_users SET final_product_score =  (product_scores_by_users.product_score - product_scores_by_users.avg_product_score) WHERE product_scores_by_users.product_score != 0"
            query_10 = "DROP TABLE IF EXISTS similarity_score_users"
            query_11 = "CREATE TABLE similarity_score_users (with_user_id INT NOT NULL,    cosine_similarity_score DECIMAL(20, 15) NOT NULL,PRIMARY KEY (with_user_id))"
            query_12 = "INSERT INTO similarity_score_users SELECT cosine_similarity_components.by_user_id, (cosine_similarity_components.numerator/(cosine_similarity_components.denominator1*cosine_similarity_components.denominator2))  as cosine_similarity_score FROM (SELECT score_for_other_user.by_user_id, SUM(score_for_first_user.final_product_score*score_for_other_user.final_product_score)  as numerator, POWER(SUM(POWER(score_for_first_user.final_product_score, 2)),0.5) as denominator1, POWER(SUM(POWER(score_for_other_user.final_product_score, 2)),0.5) as denominator2 FROM (SELECT * FROM product_scores_by_users WHERE by_user_id = %s) as score_for_first_user,   (SELECT * FROM product_scores_by_users WHERE by_user_id != %s ) as score_for_other_user WHERE score_for_first_user.for_product_id = score_for_other_user.for_product_id GROUP BY score_for_other_user.by_user_id) as cosine_similarity_components WHERE cosine_similarity_components.denominator1*cosine_similarity_components.denominator2 != 0"
            query_13 = "DROP TABLE IF EXISTS recommend_to_user_1"
            query_14 = "CREATE TABLE IF NOT EXISTS recommend_to_user_1  (for_product_id INT NOT NULL,confidence DECIMAL(20, 15) NOT NULL,PRIMARY KEY (for_product_id))"
            query_15 = "INSERT INTO recommend_to_user_1 SELECT product_scores_by_users.for_product_id, SUM(similarity_score_users.cosine_similarity_score*product_scores_by_users.final_product_score)/SUM(similarity_score_users.cosine_similarity_score) as confidence FROM similarity_score_users INNER JOIN product_scores_by_users ON similarity_score_users.with_user_id = product_scores_by_users.by_user_id GROUP BY product_scores_by_users.for_product_id"
            query_16 = "SELECT products.* FROM products INNER JOIN recommend_to_user_1 ON products.product_id = recommend_to_user_1.for_product_id  ORDER BY confidence DESC LIMIT 10"
            db_cursor = self.db_conn.cursor()       
            db_cursor.execute(query_1)               
            db_cursor.execute(query_2) 
            db_cursor.execute(query_3)
            self.db_conn.commit()
            db_cursor.execute(query_4)
            self.db_conn.commit()
            db_cursor.execute(query_5)
            self.db_conn.commit()
            db_cursor.execute(query_6)
            self.db_conn.commit()
            db_cursor.execute(query_7)
            self.db_conn.commit()
            db_cursor.execute(query_8)
            self.db_conn.commit()
            db_cursor.execute(query_9)
            self.db_conn.commit()
            db_cursor.execute(query_10)
            db_cursor.execute(query_11)
            db_cursor.execute(query_12, (user_id, user_id, ))
            self.db_conn.commit()
            db_cursor.execute(query_13)
            db_cursor.execute(query_14)
            db_cursor.execute(query_15)
            self.db_conn.commit()
            db_cursor.execute(query_16)
            column_names = [column[0] for column in db_cursor.description]        
            data = db_cursor.fetchall()        
            self.print_for_select(column_names,data)            
        except Exception as e:            
            print(e)
            sys.exit(0)
            
    def print_for_select(self, column_names,data):
        for row in data:
            pprint(dict(zip(column_names, row)))
            print()    

    def close(self):
        try:
            self.db_conn.close()
        except Exception as e:
            print(e)
            sys.exit(0)        
```
